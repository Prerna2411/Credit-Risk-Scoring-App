```python
def handle_missing_values(df, threshold=0.65):
    missing_ratio = df.isnull().mean()
    cols_to_drop = missing_ratio[missing_ratio > threshold].index

    #  Actually drop the columns
    df = df.drop(columns=cols_to_drop)
    print("Dropped columns:", list(cols_to_drop))
    # Fill numeric columns with median
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        df[col] = df[col].fillna(df[col].median())

    # Fill categorical columns with mode
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df[col].isnull().any():
            df[col] = df[col].fillna(df[col].mode()[0])

    return df
```

```python
### for tree-based models (LABEL ENCODING)
from sklearn.preprocessing import LabelEncoder

def label_encode(df):
    df=df.copy()
    label_enc=LabelEncoder()
    for col in df.select_dtypes(include='object').columns:
        df[col]=label_enc.fit_transform(df[col].astype(str))
    return df
```

```python
#for linear models (one-hot encoding)
from sklearn.preprocessing import OneHotEncoder

def one_hot(df):
    df=df.copy()
    return pd.get_dummies(df, drop_first=True)

```

```python
X=df.drop(columns=['TARGET'])
y=df['TARGET']
```

```python
X_label=label_encode(X)
X_onehot=one_hot(X)
from sklearn.model_selection import train_test_split
##split data
X_train_l,X_test_l,y_train_l,y_test_l=train_test_split(X_label,y,test_size=0.2,random_state=42)
X_train_o,X_test_o,y_train_o,y_test_o=train_test_split(X_onehot,y,test_size=0.2,random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score
from sklearn.svm import LinearSVC
```

```python
##SMOTE for one-hot encoded models
sm=SMOTE(random_state=42)
X_train_o_res,y_train_o_res=sm.fit_resample(X_train_o,y_train_o)

###class imbalance ratio
scale=y_train_l.value_counts()[0]/y_train_l.value_counts()[1]

##one-hot models
models_label={
    "Random Forest":RandomForestClassifier(n_estimators=100,class_weight='balanced',random_state=42),
    "XGBoost":XGBClassifier(use_label_encoder=False,eval_metric='logloss',scale_pos_weight=scale,random_state=42),
    "LightGBM": LGBMClassifier(class_weight='balanced', scale_pos_weight=scale, random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, scale_pos_weight=scale, random_state=42)
}


models_onehot={
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": LinearSVC(class_weight='balanced', random_state=42),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()

}

```

```python
def evaluate_models(models,X_train,X_test,y_train,y_test,title):
    print(f"\n===== Evaluating {title} Models =====")
    for name,model in models.items():
        print(f"\n▶ Training {name}...")
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        try:
            y_prob = model.predict_proba(X_test)[:, 1]
        except:
            y_prob = model.decision_function(X_test)
        print(classification_report(y_test, y_pred))
        print("ROC AUC:", roc_auc_score(y_test, y_prob))

# Run evaluations
evaluate_models(models_onehot, X_train_o, X_test_o, y_train_o, y_test_o, "One-Hot Encoded")
evaluate_models(models_label, X_train_l, X_test_l, y_train_l, y_test_l, "Label Encoded")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

cat_model = CatBoostClassifier(verbose=0, random_state=42, class_weights=[1, 5])


param_dist = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9]
}


skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


random_search = RandomizedSearchCV(
    estimator=cat_model,
    param_distributions=param_dist,
    n_iter=10,  # increase for more tuning
    scoring='roc_auc',
    cv=skf,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

X_train_l_res, y_train_l_res = SMOTE(random_state=42).fit_resample(X_train_l, y_train_l)
#X_resampled_array, y_train_l_res = SMOTE(random_state=42).fit_resample(X_train_l, y_train_l)
#X_train_l_res = pd.DataFrame(X_resampled_array, columns=X_train_l.columns)

top_features = [
    'AMT_CREDIT', 'AMT_INCOME_TOTAL', 'EXT_SOURCE_1',
    'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_EMPLOYED', 'REGION_POPULATION_RELATIVE'
]

X_train_top = X_train_l_res[top_features]
X_test_top = X_test_l[top_features]

random_search.fit(X_train_top, y_train_l_res)
```
